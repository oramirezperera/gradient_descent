## Dependencies



```python
from matplotlib import cm # for colouring
import numpy as np
import matplotlib.pyplot as plt
```

## 3D graph of the cost function

```python
fig, ax  = plt.subplots(subplot_kw={"projection":"3d"})
def f(x, y):
  return x**2 + y**2

res = 100

x = np.linspace(-4, 4, num=res)
y = np.linspace(-4, 4, num=res)

X,Y = np.meshgrid(x, y)

Z = f(X,Y)

surf = ax.plot_surface(X, Y, Z, cmap=cm.cool)

fig.colorbar(surf)
```

## Gradient Descent

```python
level_map = np.linspace(np.min(Z), np.max(Z), num= res)
plt.contourf(X, Y, Z, levels=level_map, cmap=cm.cool)
plt.colorbar()
plt.title('Gradient Descent')

p = np.random.rand(2) * 8 - 4 
plt.plot(p[0], p[1], 'o', c='k')

h = 0.01 # increment
lr = 0.01 # learning rate

def derivate(cp, p):
  return (f(cp[0], cp[1]) - f(p[0], p[1])) / h

def gradient(p):
  grad = np.zeros(2)
  for idx, val in enumerate(p):
    cp = np.copy(p)
    cp[idx] = cp[idx] + h

    dp = derivate(cp, p)
    grad[idx] = dp
  
  return grad

for i in range(1000):
  p = p - lr*gradient(p)
  if (i % 10 == 0):
    plt.plot(p[0], p[1], 'o', c='r')

plt.plot(p[0], p[1], 'o', c='w')
print(p)
```
